```python
import pandas as pd

data = pd.read_excel('Data Sampah.xlsx')

df = pd.DataFrame({
    'Kode Provinsi': data['kode_provinsi'],
    'Nama Provinsi': data['nama_provinsi'],  
    'Kode Kabupaten Kota': data['kode_kabupaten_kota'],
    'Nama Kabupaten Kota' : data['nama_kabupaten_kota'],
    'Jumlah Produksi Sampah' : data['jumlah_produksi_sampah'],
    'Satuan (Ton)' : data['satuan'],
    'Tahun' : data['tahun']
})
df.index = df.index + 1
print(df)
```

```python
# Masukkan tahun yang ingin dihitung
tahun_tertentu = 2019

# Filter data untuk Jawa Barat pada tahun tertentu
jawa_barat_data = df[(df['Nama Provinsi'] == 'Jawa Barat') & (df['Tahun'] == tahun_tertentu)]

# Hitung total produksi sampah
total_produksi_sampah = jawa_barat_data['Jumlah Produksi Sampah'].sum()

# Tampilkan hasil
print(f"Total produksi sampah di seluruh Kabupaten/Kota di Jawa Barat pada tahun {tahun_tertentu} adalah {total_produksi_sampah} Ton.")
```

```python
# Mengelompokkan data berdasarkan Tahun dan menjumlahkan produksi sampah
total_per_tahun = df.groupby('Tahun')['Jumlah Produksi Sampah'].sum()

# Menampilkan hasil
print("Total produksi sampah per tahun:")
print(total_per_tahun)
```

```python
# Mengelompokkan data berdasarkan Nama Kabupaten/Kota dan Tahun, lalu menjumlahkan produksi sampah
total_per_kota_per_tahun = df.groupby(['Nama Kabupaten Kota', 'Tahun'])['Jumlah Produksi Sampah'].sum()

# Mengatur ulang hasil agar lebih mudah dibaca
total_per_kota_per_tahun = total_per_kota_per_tahun.reset_index()

# Menampilkan hasil
print("Total produksi sampah per Kota/Kabupaten per tahun:")
print(total_per_kota_per_tahun)
```
